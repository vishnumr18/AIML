```python
import camelot
from fpdf import FPDF

# Define the input and output PDF paths
input_pdf_path = "18_unlocked.pdf"
output_pdf_path = "extracted_tables.pdf"

# Extract tables from the specified pages using Camelot
# Adjust the pages parameter as needed (e.g., '44-48' for pages 44 to 48)
tables = camelot.read_pdf(input_pdf_path, pages='44-48')

# Create a PDF document to store the extracted tables
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
pdf.set_font("Arial", size=12)

# Add each table to the PDF
for table in tables:
    pdf.add_page()
    table_data = table.df.values.tolist()
    for row in table_data:
        row_text = ' | '.join(map(str, row))
        pdf.cell(200, 10, txt=row_text, ln=True)

# Save the output PDF
pdf.output(output_pdf_path)

print(f"Extracted tables have been saved to {output_pdf_path}")
```
