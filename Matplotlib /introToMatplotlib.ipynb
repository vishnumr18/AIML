```python
hd = pd.read_csv('Data/heart-disease.csv')
hd
```

```python
hd['age'].plot.hist()
```

```python
hd.plot.hist(figsize=(10,30),subplots=True)
```

```python
over_50 = hd[hd['age']>50]
over_50
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,5))

over_50.plot(x='age',
           y='chol',
           c='target',
           ax=ax,
           kind='scatter');

```

```python
# Object Oreanted method 
fig, ax = plt.subplots(figsize=(10,6))

# Plotting the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart Disease and Cholesterol Level',
       xlabel = 'age',
       ylabel='Cholesterol Level')

ax.legend(*scatter.legend_elements(), title='Targer')

ax.axhline(y= np.mean( over_50['chol']),color='b', linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Plotting data in ax0
scatter = ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])

# customizign ax0
ax0.set(title='Age V/S Cholesterol',
        ylabel='Cholesterol')


# Plotting data in ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=np.mean(over_50['chol']), linestyle='--')

# Customizing ax1
ax1.set(title='Age V/S Thalach',
        xlabel='Age',
        ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=np.mean(over_50['thalach']), linestyle='--')

# Fixing the title for figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')
```

#### Different types of Bar Plots
    

```python
time_data = {'facebook':120, 'youtube':90, 'instagram':40, 'snapchat':28}

plt.bar(time_data.keys(), time_data.values(), width=.5, color='blue')
plt.xlabel('Social Media')
plt.ylabel('Time')
plt.title('Average Time Spent on Social Media');
```

```python
fig, ax = plt.subplots(figsize=(6,6))

color = ['royalblue', 'red', 'purple', 'yellow']

ax.bar(time_data.keys(), time_data.values(), color=color, width=.5)

ax.set(xlabel='Social Medaia',
       ylabel='Time Spent in Minutes',
       title='Average Time Spent On Social Media');
```

```python
np.random.seed(43)
fb = np.random.randint(30, 150, size=5)
yt = np.random.randint(30, 150, size=5)
ig = np.random.randint(30, 150, size=5)
sc = np.random.randint(30, 150, size=5)
fb, yt, ig, sc
```

```python
xpos = np.arange(len(fb))
xpos
```

```python
# Multiple Barchart 
fig, ax = plt.subplots(figsize=(10,6))

bwidth = .2

ax.bar(xpos, fb, color='royalblue', width= bwidth, label='FB')
ax.bar(xpos+.2, yt, color='red', width= bwidth, label='YT')
ax.bar(xpos+.4, ig, color='purple', width= bwidth, label='IG')
ax.bar(xpos+.6, sc, color='yellow', width= bwidth, label='SC')

ax.set(title="Average Time Spent on Social Media",
       xlabel='Day',
       ylabel='Time Spent in Minutes')

ax.set_xticks(xpos+.3, ('Monday', 'Tuesday', 'Wednsday', 'Thursday', 'Friday'))

ax.legend();
```

```python
# Stack Bar Chart
np.random.seed(3)
fb = np.random.randint(30, 150, size=5)
yt = np.random.randint(30, 150, size=5)
ig = np.random.randint(30, 150, size=5)
sc = np.random.randint(30, 150, size=5)

x = ['Monday', 'Tuesday', 'Wednsday', 'Thursday', 'Friday']
print(fb, yt, ig, sc)
```

```python
xpos = np.arange(len(x))
bwidth=.3

fig, ax = plt.subplots(figsize=(10,8))

ax.bar(xpos, fb, color='royalblue', width=bwidth, label= "FB")
ax.bar(xpos, yt, bottom=fb, color='red', width=bwidth, label= "YT")
ax.bar(xpos, ig, bottom=yt, color='purple', width=bwidth, label= "IG")
ax.bar(xpos, sc, bottom=ig, color='yellow', width=bwidth, label= "SC")

ax.legend()
ax.set_xticks(xpos, x);
```

```python
xpos = np.arange(len(x))
bwidth=.3

fig, ax = plt.subplots(figsize=(10,8))

ax.barh(xpos, fb, color='royalblue', height=.5, label= "FB")
ax.barh(xpos, yt, left=fb, color='red', height=.5, label= "YT")
ax.barh(xpos, sc, left=ig, color='yellow', height=.5, label= "SC")
ax.barh(xpos, ig, left=yt, color='purple', height=.5, label= "IG")

ax.set(title='Time Spent On Social Media',
       ylabel='Day',
       xlabel='Time Spent in Minutes')

ax.legend()
ax.set_yticks(xpos, x);
```
