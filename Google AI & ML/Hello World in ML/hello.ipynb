```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

    tf.keras.Sequential: This creates a sequential model, which is a linear stack of layers. It means the layers are stacked on top of each other, and the data flows sequentially from one layer to the next.

    [keras.layers.Dense(units=1, input_shape=[1])]: This line adds a dense layer to the model. A dense layer is a fully connected layer, meaning each neuron in the layer is connected to every neuron in the previous layer. Here's what the parameters mean:
        * units=1: This specifies the number of neurons in the layer. In this case, there is only one neuron. Since there's only one neuron and no activation function specified, this layer will output a single numerical value.
        * input_shape=[1]: This specifies the shape of the input data. In this case, it's a one-dimensional input with a single feature. The input shape [1] means the model expects input data to be a one-dimensional array with one element.

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])]);
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0], dtype=float)
ys = np.array([-2.0,1.0,4.0,7.0,10.0,13.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```
